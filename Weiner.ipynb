```python
no=2000
D=np.linspace(0,4*np.pi,no)
D=np.sin(D)
D=D/np.max(D)
noise=np.random.normal(0,0.3,no)
signl=D+noise
```

```python
plt.plot(D)
```

```python
plt.plot(noise)
```

```python
plt.plot(signl)
```

```python
M=10
n=35
out=list()
e=list()
wts=list()
for i in range(M,len(signl)-n):
    ls=list()
    for j in range(n,0,-1):
        ls.append(signl[i+j-M:i+j][::-1])
    u=np.matrix(ls).T
    R=u*u.T
    dn=np.matrix(D[i:i+n]).T
    p=u*dn
    inv=np.linalg.inv(R)
    w0=inv*p
    vl=w0.T*u.T[0].T
    wts.append(np.sum(w0))
    out.append(np.array(vl)[0][0])
    e.append((np.array(dn[-1]-vl)[0][0])**2)
```

```python
plt.plot(out)
```

```python
plt.plot(signl)
```

```python
plt.plot(e)
```

```python
plt.plot(wts)
```
